```python
# Começa com a string
first = times[0]

#Extrai as partes dos componentes, primeiro os minutos
minutes, rest = first.split(':')

#agora do resto tira os segundos e centésimos
seconds, centis = rest.split('.')

# converter os tempos para inteiros
convert_time = (int(minutes) * 60 * 100) + (int(seconds) * 100) + int(centis)

print(convert_time)
```

```python
print("USANDO FOR")
for time in times:
    print(time)
    
##isso só que com while
print("USANDO WHILE")
index = 0
while index < len(times):
    print(times[index])
    index += 1
```

```python
#combinando os dois para listar os tempos convertidos
for time in times:
    minutes, rest = time.split(':')
    seconds, centis = rest.split('.')
    converted_time = (int(minutes) * 60 * 100) + (int(seconds) * 100) + int(centis)
    print(time, '->', converted_time)
```
