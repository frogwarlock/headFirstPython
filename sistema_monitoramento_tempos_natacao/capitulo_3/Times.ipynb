O objetivo é trazer o resultado dos dois notebooks, então as variávies da tarefa um, a lista de tempos e a média calculada na tarefa dois.

```python
FN = "Darius-13-100m-Fly.txt"
```

```python
swimmer, age, distance, stroke = FN.removesuffix('.txt').split('-')
```

```python
FOLDER = "../swimdata/"
```

```python
with open(FOLDER + FN) as file:
    lines = file.readlines()
    times = lines[0].strip().split(',')
```

```python
converted_times = []
for time in times:
    minutes, rest = time.split(':')
    seconds, centis = rest.split('.')
    converted_times.append((int(minutes) * 60 * 100) + (int(seconds) * 100) + int(centis))
```

```python
import statistics
average = statistics.mean(converted_times)
```

```python
mins_secs, centis = str(round(average/100, 2)).split('.')
mins_secs = int(mins_secs)
minutes = mins_secs // 60
seconds = mins_secs - minutes * 60
```

```python
print(f"Swimmer: {swimmer}, Age: {age}, Distance: {distance}, Stroke: {stroke}")
print('Tempos:')
for time in times:
    print(f"- {time}")
    
print(f"\nAverage time: {minutes}:{seconds:02}.{centis}")
```
